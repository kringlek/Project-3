```python
#Import Dependencies
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
from datetime import datetime as dt
import datetime
```

```python
#Read in Data from CSV
hurricane_csv = "Data/hurricane_info.csv"
hurricane_data = pd.read_csv(hurricane_csv)
hurricane_data.head()
```

```python
#Delete ref column
hurricane_data.pop('REf')

#Filter out unwanted rows/names (nondescrptive hurricane names such as "four", not specific enough/ not usable)
hurricane_data = hurricane_data.loc[(hurricane_data["Name"]!="Unnamed") & (hurricane_data["Name"]!="Four") & (hurricane_data["Name"]!="Five") & (hurricane_data["Name"]!="Two") & (hurricane_data["Name"]!="Eight") & (hurricane_data["Name"]!="Seven") & (hurricane_data["Name"]!="Fifteen") & (hurricane_data["Name"]!="Thirteen") & (hurricane_data["Name"]!="Three") & (hurricane_data["Name"]!="One")& (hurricane_data["Name"]!="Six")& (hurricane_data["Name"]!="Nine") & (hurricane_data["Name"]!="Ten")& (hurricane_data["Name"]!="Eleven") & (hurricane_data["Name"]!="Fourteen")  & (hurricane_data["Name"]!="Item")  & (hurricane_data["Name"]!="How") & (hurricane_data["Name"]!="Twelve") & (hurricane_data["Name"]!="Eighteen")]
```

```python
#Drop na columns
hurricane_data.dropna(inplace = True)

#year Column
date_test = hurricane_data['Duration'].str.split('-', expand=True)
date_test_year = date_test[1].str.split(',',expand=True)
hurricane_data['Year'] = date_test_year[1]
```

```python
#Separate month/day columns
#extract start date only
test = hurricane_data['Duration'].str.split('-', expand=True)
hurricane_data['Start Date']  = test[0]

#Separate Columns for month/day
new_date = hurricane_data['Start Date'].str.split(expand=True)
hurricane_data["Month"]= new_date[0]
hurricane_data["Day"]= new_date[1]
hurricane_data
```

```python
#Separate Columns for Wind Speed / Pressure values & units
new_wind = hurricane_data['Wind speed'].str.split(expand=True)
hurricane_data['Wind Speed (mph)'] = new_wind[0]
#hurricane_data['Wind Speed'] = new[0]
#hurricane_data['Wind Speed Units'] = new[1]

new_press = hurricane_data['Pressure'].str.split(expand=True)
hurricane_data['Pressure (hPa)'] = new_press[0]
#hurricane_data['Pressure'] = new[0]
#hurricane_data['Pressure Units'] = new[1]

#Drop na columns again now that date is separated
hurricane_data.dropna(inplace = True)
```

```python
#Filter out unwanted rows/names for Damages Column
hurricane_data = hurricane_data.loc[(hurricane_data["Damage"]!="Extensive") & (hurricane_data["Damage"]!="[11][100][101][102]") & (hurricane_data["Damage"]!="[108][109]") & (hurricane_data["Damage"]!="[131][132][133]")  & (hurricane_data["Damage"]!="Minimal")  & (hurricane_data["Damage"]!="[142]")  & (hurricane_data["Damage"]!="[159][160][161][162][163][164][165][166][167]")  & (hurricane_data["Damage"]!="[144]")  & (hurricane_data["Damage"]!="[174]") & (hurricane_data["Damage"]!="Unknown")  & (hurricane_data["Damage"]!="Heavy") & (hurricane_data["Damage"]!="Minor")  & (hurricane_data["Damage"]!="Moderate") & (hurricane_data["Damage"]!="Millions") & (hurricane_data["Damage"]!="13") & (hurricane_data["Damage"]!="5") & (hurricane_data["Damage"]!="None")]

#Separate Damages Cost
new_cost = hurricane_data['Damage'].str.split(expand=True)
hurricane_data['Value'] = new_cost[0]
hurricane_data['Amount'] =new_cost[1]

cost_test = hurricane_data['Value'].str.split('$',expand=True)
hurricane_data['Damages Cost ($)'] = cost_test[1]
hurricane_data
```

```python
#Adjust format of df
hurricane_data['Damages Cost ($)'] = hurricane_data['Damages Cost ($)'].str.replace(',', '').astype(float)
hurricane_data.dtypes
```

```python
#Loop through damages - millions, billions, or grab thousands (normal format)
def sorting(row):
    if row['Amount'] == 'million':
        var = row['Damages Cost ($)'] * (1000000)
        return var
    elif row['Amount'] == 'billion':
        var = row['Damages Cost ($)'] * (1000000000)
        return var
    elif row['Amount'] == 'thousand':
        var = row['Damages Cost ($)'] * (1000)
        return var
hurricane_data['Damages ($)'] = hurricane_data.apply(lambda row: sorting(row), axis=1)
hurricane_data
```

```python
#Analyze Data by different Categories
#sort by damages cost, deaths
damages_df = hurricane_data.sort_values("Damages Cost ($)", ascending=False)
deaths_df = hurricane_data.sort_values("Deaths", ascending=False)

#group by year, hurricane category 
hurricane_category_df = hurricane_data.groupby(["Year", "Category"])

#group by areas affected, damages cost
location_damages_df = hurricane_data.groupby(["Areas affected", "Damages Cost ($)"])
```

```python
#Remove if death number is unknown
hurricane_data = hurricane_data.loc[(hurricane_data["Deaths"]!="Unknown")]

#Clean Table to Columns Wanted (a lot of extra columns not needed from data cleaning in one df)
hurricane_data_clean = hurricane_data[['Name', 'Areas affected', 'Month', 'Day', 'Year', 'Wind Speed (mph)', 'Pressure (hPa)', 'Damages ($)','Deaths' ]].copy()

#Drop na columns again now that date is separated
hurricane_data_clean.dropna(inplace = True)
hurricane_data_clean
```

```python
#Write to csv
hurricane_data_clean.to_csv("Data/hurricane_clean_dataset.csv",index=False)
```
